```python
import pandas as pd
# Load the CSV file
file_path = "/Users/manoharshasappa/Desktop/Files/Github_Files/notable_ai_models.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic info about the DataFrame
print("DataFrame Info:")
print(df.info())
```

```python
# Display first few rows for reference
print("\nFirst 5 Rows of the DataFrame:")
print(df.head(5))
```

```python
# Check for null values
print("\nNull Values in Each Column:")
print(df.isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Year'] = pd.to_datetime(df['Publication date'], errors='coerce').dt.year

#Number of models per year
models_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.bar(models_per_year.index, models_per_year.values, color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Models')
plt.title('Number of AI Models Released Per Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Total number of models in dataset
unique_models = df['System'].dropna().unique()
num_models = len(unique_models)

# Print the number of unique AI models and list them
print(f"Total Unique AI Models: {num_models}")
print("List of AI Models:")
for model in unique_models:
    print(model)
```

```python
# Extract years from the 'Publication date' column
df['Year'] = pd.to_datetime(df['Publication date'], errors='coerce').dt.year

# Number of models per organization
models_per_org = df['Organization'].value_counts().head(10) 
plt.figure(figsize=(12, 6))
plt.barh(models_per_org.index, models_per_org.values, color='purple')
plt.xlabel('Number of Models')
plt.ylabel('Organization')
plt.title('Top 10 Organizations by AI Model Count')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Graph: Distribution of models by domain
models_by_domain = df['Domain'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(models_by_domain.index, models_by_domain.values, color='orange')
plt.xlabel('Domain')
plt.ylabel('Number of Models')
plt.title('Distribution of AI Models by Domain')
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# AI Model Growth Over the Years
models_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.plot(models_per_year.index, models_per_year.values, marker='o', linestyle='-', color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Models')
plt.title('AI Model Growth Over the Years')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Model accessibility distribution
model_accessibility = df['Model accessibility'].value_counts()
plt.figure(figsize=(12, 6))
plt.pie(model_accessibility, labels=model_accessibility.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'orange', 'red', 'purple'])
plt.title('AI Model Accessibility Distribution')
plt.show()
```
